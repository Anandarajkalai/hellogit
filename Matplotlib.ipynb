```python
from numpy import linspace,log
from scipy.misc import factorial

complexity=DataFrame(index=linspace(0,100,1000)[1:])

complexity['constant']=1
complexity['linear']=linspace(0,100,1000)[1:]
complexity['quadratic']=complexity['linear']**2
complexity['cubic']=complexity['linear']**3
complexity['log']=log(complexity['linear'])
complexity['factorial']=factorial(complexity['linear'])
complexity['exponential']=2**complexity['linear']

complexity.head()
```

```python
complexity.plot()
```

```python
complexity[:2.5].plot()
```

```python
complexity[['log','linear','quadratic']].plot()
```

```python
from matplotlib.pyplot import plot,axes

plot(complexity.index,complexity['log'],'g-',label='Log')
plot(complexity.index,complexity['quadratic'],'b-',label='Quadratic')

axes().set_xlim((0,10))
axes().set_ylim((0,25))

axes().legend(loc='upper right')
```

```python
from matplotlib.pyplot import subplots

fig,axes=subplots(nrows=1,ncols=2, figsize=(15,5))

axes[0].plot(complexity.index,complexity['log'],'g-',label='Log')
axes[0].plot(complexity.index,complexity['quadratic'],'b-',label='Quadratic')
axes[0].legend(loc='center right')

axes[1].plot(complexity.index,complexity['log'],'g-',label='Log')
axes[1].plot(complexity.index,complexity['linear'],'b-',label='Linear')
axes[1].legend(loc='upper left')
```
